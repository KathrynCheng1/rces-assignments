### 5) Extract the state or country using Pandas [text data functions](https://pandas.pydata.org/pandas-docs/stable/text.html)

Add it as a new column to the dataframe called `country`. Note that some of the "countries" are actually U.S. states.

```python
df = pd.read_csv('http://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv', parse_dates=['time'], index_col='id')
df['country'] = df.place.str.split(', ').str[-1]
df.head()
```

```python
df.head()
```

```python
unique_countries = df['country'].unique()
print(unique_countries)
```

### 7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger and

```python
filtered_df = df[df['mag'] >= 4]
filtered_df.head()
```

### 8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes

Location name on the x axis, Earthquake count on the y axis

```python
earthquake = filtered_df['country'].value_counts()
top5_loc = earthquake.head(5)

plt.figure(figsize=(10, 6))
top5_loc.plot(kind='bar')
plt.title('Top 5 Locations with the Most Earthquakes and Magnitude is Greater than or Equal to 4')
plt.xlabel('Country/State')
plt.ylabel('Earthquake Count')
plt.tight_layout()
plt.show()
```

### 9) Make a histogram the distribution of the Earthquake magnitudes

https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.DataFrame.hist.html
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html

Do one subplot for the filtered and one for the unfiltered dataset.
Use a Logarithmic scale. What sort of relationship do you see?

```python
fig, axes = plt.subplots(figsize=(12, 5), ncols=2, width_ratios=[2, 2])
ax0, ax1 = axes

mag_filtered = filtered_df['mag']
ax0.hist(mag_filtered, bins=4, edgecolor='black')
ax0.set_title('Histogram of Earthquake Magnitudes for Filtered Data')
ax0.set_xlabel('Magnitude')
ax0.set_ylabel('Frequency')
ax0.set_yscale('log')

mag = df['mag']
ax1.hist(mag, bins=8, edgecolor='black')
ax1.set_title('Histogram of Earthquake Magnitudes for Non-Filtered Data')
ax1.set_xlabel('Magnitude')
ax1.set_ylabel('Frequency')
ax1.set_yscale('log')
plt.show()

print("The relationship is that there are more earthquakes at lower magnitude in both the filtered and non-filtered data")
```
